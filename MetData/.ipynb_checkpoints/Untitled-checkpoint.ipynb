```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
lp=pd.read_csv('lipidPanel.csv')
lp.replace(['test could not be performed','Test could not be performed','NaN'],value='NaN')
```

```python
cols= lp.columns
for col in cols:
    item=pd.to_numeric(lp[col],errors='coerce')
    pval=stats.normaltest(item,nan_policy='omit').pvalue
    if pval < .05:
        print col+' Normal'
    else:
        logged=item.apply(np.log)
        pval2=stats.normaltest(logged,nan_policy='omit').pvalue
        if pval2 < .05:
            print col+'Log Normal'
        
        
```
